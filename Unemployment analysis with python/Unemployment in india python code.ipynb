```python
plt.figure(figsize=(10,5))
sns.lineplot(x=' Date', y=' Estimated Unemployment Rate (%)', data=data, color='orange')
plt.title("ðŸ“‰ Unemployment Rate Over Time in India", fontsize=14)
plt.xlabel("Date")
plt.ylabel("Unemployment Rate (%)")
plt.xticks(rotation=90)
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='Region', y=' Estimated Unemployment Rate (%)', data=data, palette='viridis')
plt.title("ðŸ“Š Region-wise Unemployment Rate", fontsize=14)
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.heatmap(
    data[[' Estimated Unemployment Rate (%)', 
          ' Estimated Employed', 
          ' Estimated Labour Participation Rate (%)']].corr(),
    annot=True, cmap='coolwarm'
)
plt.title("ðŸ” Correlation Heatmap", fontsize=13)
plt.show()
```
